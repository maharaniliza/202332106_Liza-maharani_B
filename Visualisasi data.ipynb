## MEMBUAT BAR CHART

```python
MataKuliah = ['Dasar Pemrograman', 'Basis Data', 'Machine Learning', 'Pemrograman Objek', 'Pemrograman Web']
Jumlah_mahasiswa = [125, 90, 120, 99, 77]

plt.figure(figsize=(22, 7))
plt.barh(MataKuliah, Jumlah_mahasiswa, color='blue')
plt.title('Jumlah Mahasiswa pada matakuliah di semester 3', size=14)
plt.xlabel('Jumlah Mahasiswa', size=12)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```

## PIE CHART

```python
import matplotlib.pyplot as plt

# Data produk dan jumlah terjual
stock = ['Cheese', 'Bread', 'Milk', 'Ice Cream', 'Noodles', 'Egg']
sold = [10, 50, 45, 15, 37, 113]

# Membuat pie chart
fig = plt.figure()
ax = fig.add_subplot(111) # Menggunakan add_subplot untuk menambahkan axes ke figure
ax.axis('equal')  # Memastikan pie chart berbentuk lingkaran
ax.pie(sold, labels=stock, autopct='%1.2f%%') # Memperbaiki autopct

# Mengatur aspek rasio agar tidak ada peringatan
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

```python
akomodasi = [2000, 1100, 4180, 5800, 6420, 980]
kamar = [1503, 10732, 3555, 5206, 2187, 4731, 522]
kasur = [9257, 16415, 6045, 7832, 2885, 8080, 786]

fig, ax = plt.subplots(figsize=(8, 6))  # Ukuran figure yang lebih jelas

box = ax.boxplot([akomodasi, kamar, kasur], vert=False, showmeans=True, meanline=True, tick_labels=['Akomodasi', 'Kamar', 'Kasur'], patch_artist=True) # Membuat boxplot horizontal, menampilkan mean sebagai garis

medianprops = {'linewidth': 2, 'color': 'blue'}  # Warna garis median
meanprops = {'linewidth': 2, 'color': 'red'}     # Warna garis mean

colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

ax.set_aspect('auto') # Menyesuaikan aspek agar tidak terjadi peringatan

plt.show()
```

## SCATTER PLOT

```python
import matplotlib.pyplot as plt

# Data nilai tugas, UTS, dan UAS
tugas = [58, 80, 65, 79, 82]
uts = [78, 80, 85, 59, 92]
uas = [62, 79, 87, 77, 84]

fig, ax = plt.subplots()

# Scatter plot hubungan tugas dengan UTS
ax.scatter(tugas, uts, color='red', label='Tugas vs UTS')
# Scatter plot hubungan tugas dengan UAS
ax.scatter(tugas, uas, color='blue', label='Tugas vs UAS')

# Menambahkan Label sumbu dan judul
tempX = 'Tugas'
tempY = 'Nilai'
ax.set_xlabel(tempX)
ax.set_ylabel(tempY)
ax.set_title('Scatter Plot')

# Menampilkan Legenda
ax.legend()

# Menampilkan plot
plt.show()
```
